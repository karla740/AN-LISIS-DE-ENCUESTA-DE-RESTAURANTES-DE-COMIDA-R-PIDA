# Estructura del proyecto

Este proyecto está diseñado para leer un archivo CSV con información de encuestas y convertirlo en una estructura organizada en Python para poder analizar los datos fácilmente.


### 1. Importación de librerías
Se utiliza la librería `csv`, que permite leer archivos en formato CSV (hojas de cálculo o tablas de datos). Esta librería facilita convertir cada fila del archivo en datos que Python puede procesar.


### 2. Función `cargar_datos`
Esta función es la encargada de leer el archivo y organizar la información.

Recibe como parámetro el nombre del archivo y realiza los siguientes pasos:
- Crea una lista vacía donde se guardarán los encuestados.
- Abre el archivo CSV en modo lectura.
- Usa `csv.DictReader` para leer cada fila como un diccionario.
- Recorre cada fila y la transforma en una estructura más organizada.


### 3. Estructura de los datos
Cada encuestado se guarda como un diccionario con varias secciones:

- **id:** identificador del encuestado.
- **preferencias:** contiene la comida preferida y la frecuencia de consumo.
- **consumo:** almacena el gasto promedio.
- **experiencia:** guarda la satisfacción del producto, servicio, tiempo de entrega y percepción de precios.
- **nps:** contiene datos de recomendación, si volvería a comprar y la calificación general.

Esta estructura permite agrupar la información de forma ordenada y fácil de analizar en los reportes.


### 4. Manejo de errores
El programa usa `try-except` para evitar errores si el archivo no existe. Si no se encuentra el archivo, muestra un mensaje de error en pantalla.


### 5. Carga de datos
Finalmente, se define la ruta del archivo:

```python
ruta = "encuesta_restaurante.csv"

````python
import csv

def cargar_datos(nombre_archivo):
    lista_encuestados = []
    try:
        with open(nombre_archivo, mode='r', encoding='utf-8') as archivo:
            lector = csv.DictReader(archivo)
            for fila in lector:
                registro = {
                    "id": int(fila["id"]),
                    "preferencias": {
                        "comida_preferida": fila["comida_preferida"],
                        "frecuencia": fila["frecuencia_consumo"],
                    },
                    "consumo":{
                        "gasto": float(fila["gasto_promedio"])
                    },
                    "experiencia": {
                        "producto": int(fila["satisfaccion_producto"]),
                        "servicio": int(fila["satisfaccion_servicio"]),
                        "tiempo de entrega": fila["tiempo_entrega"],
                        "precio": fila["precio_percepcion"],
                    },
                    "nps": {
                        "recomendacion": int(fila["recomendaria"]),
                        "volveria": fila["volveria_comprar"].strip().lower() == "true",
                        "general": int(fila["calificacion_general"])
                    }
                }
                lista_encuestados.append(registro)
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo {nombre_archivo}")
    return lista_encuestados

#Aqui almacenamos la ruta del archivo
ruta="encuesta_restaurante.csv" 

#Mandamos a llamar la funsion que lee el archivo y donde la esta estructura
datos=cargar_datos(ruta) 
````

# Reporte 1. Cantidad de personas por comida preferida

## ¿Qué ponemos en práctica?
- Frecuencia de datos: Uso de diccionarios para el conteo de categorías cualitativas.  
- Manejo de diccionarios: Aplicación del método de búsqueda de claves para actualizar contadores. 


## Lógica del programa
El algoritmo inicializa un diccionario vacío. Al iterar sobre la lista de registros, accede a la clave anidada `preferencias -> comida_preferida`.  
Si la comida ya existe como clave en el acumulador, incrementa su valor en 1; de lo contrario, crea la clave con un valor inicial de 1.  
Esto permite procesar cualquier cantidad de categorías sin conocerlas de antemano.  


````python
def personas_por_comida(datos):
    # Inicializamos un diccionario vacío para ir guardando los resultados
    conteo_comidas = {}
    
    # Recorremos la lista completa de encuestados, persona por persona
    for p in datos:
        # Aseguramos que la sección "preferencias" exista antes de intentar acceder a ella
        if "preferencias" in p:
            # Extraemos el valor de la comida preferida de esa persona
            comida = p["preferencias"]["comida_preferida"]
            
            # Si la comida ya está registrada en nuestro diccionario, simplemente le sumamos 1 a su contador
            if comida in conteo_comidas:
                conteo_comidas[comida] += 1
            # Si la comida aún no está, la agregamos al diccionario con un valor inicial de 1
            else:
                conteo_comidas[comida] = 1
                
    # Retornamos el diccionario completo con el conteo final de todas las comidas
    return conteo_comidas

# Imprimiendo resultado
print("\n--- 1. Personas por comida preferida ---")
for comida, cantidad in personas_por_comida(datos).items():
                print(f"   - {comida}: {cantidad} personas")
````

## Conclusión
Este reporte permite identificar el producto estrella del restaurante. Una alta concentración en una categoría específica sugiere dónde deben enfocarse los esfuerzos de inventario y marketing.  


# Reporte 2. Frecuencia de consumo

## ¿Qué ponemos en práctica?
- Modelado de comportamiento: Organización de información basada en hábitos recurrentes.  
- Estructuras anidadas: Acceso y procesamiento de niveles internos dentro de un diccionario.  

## Lógica del programa
Se recorre la lista principal para localizar la clave `frecuencia` dentro del bloque de `preferencias`.  
El programa utiliza una lógica de clasificación donde cada respuesta (ej. "Semanal", "Diario") se convierte en una clave del diccionario de resultados.  
Si la clave ya está presente, se actualiza el valor; si no, se inserta una nueva entrada.  


````python
def frecuencia_consumo(datos):
    # Creamos un diccionario vacío para contar las respuestas
    conteo_frecuencia = {}
    
    # Pasamos por cada uno de los registros de nuestra base de datos
    for p in datos:
        # Verificamos que los datos de "preferencias" vengan en el registro
        if "preferencias" in p:
            # Sacamos el dato de la frecuencia (ej. "Semanal", "Mensual")
            frecuencia = p["preferencias"]["frecuencia"]
            
            # Si ya vimos esta frecuencia antes, aumentamos su contador en 1
            if frecuencia in conteo_frecuencia:
                conteo_frecuencia[frecuencia] += 1
            # Si es la primera vez que vemos esta frecuencia, la registramos empezando en 1
            else:
                conteo_frecuencia[frecuencia] = 1
                
    # Devolvemos nuestro diccionario ya con los totales calculados
    return conteo_frecuencia

# Imprimiendo resultado
print("\n--- 2. Frecuencia de consumo ---")
for frecuencia, cantidad in frecuencia_consumo(datos).items():
                print(f"   - {frecuencia}: {cantidad} personas")
````

## Conclusión
Determinar la frecuencia de consumo ayuda a segmentar a los clientes según su lealtad.  
Estos datos son fundamentales para diseñar estrategias de retención para aquellos que consumen con menor regularidad.  


# Reporte 3. Promedio de gasto

## ¿Qué ponemos en práctica?
- Acumuladores numéricos: Uso de variables para la suma total de valores financieros.  
- Aritmética de datos: Implementación de fórmulas para el cálculo de medias y uso de funciones de redondeo.  
- Programación defensiva: Validación de la existencia de datos numéricos y control de división por cero.  

## Lógica del programa
El programa extrae el valor numérico ubicado en la ruta `consumo -> gasto` de cada encuestado.  
Se acumula el total en una variable y, tras completar la iteración, se divide entre el total de registros obtenidos mediante `len()`.  
El resultado final se redondea para presentar una cifra coherente con un formato monetario.  


````python
def promedio_gasto(datos):
    # Creamos una variable en 0 donde iremos sumando todo el dinero gastado
    suma_gasto = 0
    # Obtenemos la cantidad exacta de personas midiendo el tamaño de la lista
    total_personas = len(datos)
    
    # Si la lista está vacía (0 personas), regresamos 0 de inmediato 
    # para evitar que el programa falle al intentar dividir entre cero
    if total_personas == 0:
        return 0
        
    # Recorremos a cada persona en la encuesta
    for p in datos:
        # Nos aseguramos de que el bloque "consumo" exista en sus datos
        if "consumo" in p:
            # Sumamos el gasto de esta persona a nuestro total acumulado
            suma_gasto += p["consumo"]["gasto"]
            
    # Calculamos el promedio matemático: todo lo sumado dividido entre el total de gente
    promedio = suma_gasto / total_personas
    
    # Retornamos el resultado, pero redondeado a solo 2 decimales para que se vea como dinero
    return round(promedio, 2)

# Imprimiendo resultado
print(f"\n--- 3. Promedio de gasto ---\n   $ {promedio_gasto(datos)}")
````

## Conclusión
El gasto promedio es el indicador clave para entender el "ticket promedio" del negocio.  
Un incremento en este valor impacta directamente en la rentabilidad general.  


# Reporte 4. Promedio de satisfacción del producto

## ¿Qué ponemos en práctica?
- Métricas de satisfacción: Procesamiento de escalas numéricas (0 a 10) para medir calidad percibida.  
- Extracción selectiva: Navegación hacia bloques específicos de experiencia del usuario.  

## Lógica del programa
El algoritmo se enfoca en el bloque `experiencia` para obtener el valor de la clave `producto`.  
Se realiza una sumatoria de todas las calificaciones individuales.  
Al finalizar el recorrido, se calcula el promedio aritmético.  
Esta métrica refleja la aceptación técnica y organoléptica de los alimentos ofrecidos.  


````python
def promedio_satisfaccion_producto(datos):
    # Iniciamos nuestro acumulador en 0 para ir sumando las notas
    suma_satisfaccion = 0
    # Contamos cuántas personas respondieron la encuesta en total
    total_personas = len(datos)
    
    # Prevenimos un error del sistema asegurándonos de no dividir entre cero
    if total_personas == 0:
        return 0
        
    # Revisamos los datos persona por persona
    for p in datos:
        # Comprobamos que el diccionario tenga la información de su "experiencia"
        if "experiencia" in p:
            # Tomamos la nota que le dio al producto y la sumamos al gran total
            suma_satisfaccion += p["experiencia"]["producto"]
            
    # Sacamos el promedio dividiendo el puntaje total entre todos los encuestados
    promedio = suma_satisfaccion / total_personas
    
    # Devolvemos el número redondeado a dos decimales para mayor claridad
    return round(promedio, 2)

# Imprimiendo resultado
print(f"\n--- 4. Satisfacción del producto ---\n   {promedio_satisfaccion_producto(datos)} / 10 pts")
````

## Conclusión
Este indicador mide la calidad del producto final.  
Un promedio alto valida los procesos de cocina, mientras que uno bajo sugiere una revisión inmediata de recetas o ingredientes.  


# Reporte 5. Promedio de satisfacción del servicio

## ¿Qué ponemos en práctica?
- Análisis de servicio: Evaluación de factores humanos y operativos en la atención.  
- Manejo masivo de datos: Procesamiento eficiente de grandes volúmenes de registros para generar un indicador de desempeño.  

## Lógica del programa
De manera similar al reporte anterior, el programa navega hacia `experiencia -> servicio`.  
Se suma la calificación otorgada por cada cliente a la atención recibida y se divide por el volumen total de encuestados.  
La lógica separa la percepción del producto de la percepción de la atención para diagnósticos más precisos.  


````python
def promedio_satisfaccion_servicio(datos):
    # Variable donde iremos sumando todas las calificaciones del servicio
    suma_satisfaccion = 0
    # Usamos len() para saber el total de personas en la lista
    total_personas = len(datos)
    
    # Si no hay nadie registrado, devolvemos 0 como seguridad
    if total_personas == 0:
        return 0
        
    # Iteramos sobre todos los clientes encuestados
    for p in datos:
        # Validamos que la pregunta sobre su "experiencia" sí fue contestada
        if "experiencia" in p:
            # Añadimos la calificación del servicio al acumulador
            suma_satisfaccion += p["experiencia"]["servicio"]
            
    # Calculamos la media (suma total de calificaciones / cantidad de personas)
    promedio = suma_satisfaccion / total_personas
    
    # Entregamos el promedio final redondeado a dos cifras
    return round(promedio, 2)

# Imprimiendo resultado
print(f"\n--- 5. Satisfacción del servicio ---\n   {promedio_satisfaccion_servicio(datos)} / 10 pts")
````

## Conclusión
La satisfacción del servicio es vital para la experiencia general.  
Identificar brechas entre la calidad del producto y la calidad del servicio permite enfocar capacitaciones para el personal de atención al cliente.  



# Reporte 6. Distribución del tiempo de entrega

## ¿Qué ponemos en práctica?
- Manejo de variables categóricas: Clasificación de opiniones subjetivas (ej. "Rápido", "Lento") en un inventario de frecuencias.  
- Contadores dinámicos: Uso de diccionarios para agrupar datos sin conocer las etiquetas de respuesta de antemano.  

## Lógica del programa
El algoritmo recorre la lista principal y accede al bloque anidado `experiencia` para extraer la opinión sobre el tiempo de entrega.  
Utiliza una lógica de búsqueda: si la categoría ya existe en el diccionario de conteo, incrementa el valor; si no, la crea con un valor inicial de 1.  
Esto permite ver cuántas personas seleccionaron cada opción de tiempo disponible en la encuesta.  


````python
def distribucion_tiempo(datos):
    # Inicializamos un diccionario vacío para contar las respuestas de tiempo
    conteo_tiempo = {}
    
    # Recorremos a cada persona encuestada en nuestra lista
    for p in datos:
        # Verificamos que la información de "experiencia" exista para evitar errores
        if "experiencia" in p:
            # Extraemos lo que opinó sobre el tiempo de entrega (ej. "Rápido", "Lento")
            tiempo = p["experiencia"]["tiempo de entrega"]
            
            # Si ya tenemos esa respuesta en el diccionario, le sumamos 1
            if tiempo in conteo_tiempo:
                conteo_tiempo[tiempo] += 1
            # Si es una respuesta nueva, la registramos con el valor de 1
            else:
                conteo_tiempo[tiempo] = 1
                
    # Retornamos el diccionario con el conteo total
    return conteo_tiempo

#Imprimiendo reultados
tiempos = distribucion_tiempo(datos)
for tiempo, cantidad in tiempos.items():
     print(f"   - {tiempo}: {cantidad} respuestas")
````

## Conclusión
Este reporte es un indicador crítico de la eficiencia logística.  
Una alta frecuencia en la categoría "Lento" sugiere que, aunque la comida sea buena, el restaurante está perdiendo clientes por demoras operativas.  


# Reporte 7. Distribución de percepción de precios

## ¿Qué ponemos en práctica?
- Análisis de sentimiento económico: Procesamiento de cómo el cliente valora el costo del producto.  
- Uso de `.items()` y `for`: Recorrido eficiente de estructuras de datos para generar resúmenes estadísticos.  

## Lógica del programa
Similar al reporte anterior, el programa busca en el diccionario de cada encuestado la clave `experiencia -> precio`.  
El objetivo es cuantificar las opiniones sobre si el precio es percibido como "Caro", "Justo" o "Barato".  
El resultado final es un diccionario que muestra la distribución total de estas percepciones entre los 10,000 registros.  


````python
def distribucion_precios(datos):
    # Creamos un diccionario vacío para ir guardando cómo perciben los precios
    conteo_precios = {}
    
    # Pasamos por cada registro de la base de datos
    for p in datos:
        # Aseguramos que el bloque "experiencia" esté presente
        if "experiencia" in p:
            # Sacamos el dato del precio (ej. "Medio", "Caro")
            precio = p["experiencia"]["precio"]
            
            # Si esta percepción ya existe, aumentamos su contador
            if precio in conteo_precios:
                conteo_precios[precio] += 1
            # Si no existe aún, la agregamos empezando en 1
            else:
                conteo_precios[precio] = 1
                
    # Devolvemos el diccionario ya lleno con los resultados
    return conteo_precios

#Imprimiendo resultados
precios = distribucion_precios(datos)
for precio, cantidad in precios.items():
                print(f"   - {precio}: {cantidad} respuestas")
````


## Conclusión
La percepción de precios es fundamental para la estrategia de pricing.  
Si la mayoría percibe el precio como "Caro" pero la satisfacción es baja, es necesario ajustar los precios o aumentar el valor percibido del producto.  


# Reporte 8. Promedio general de satisfacción

## ¿Qué ponemos en práctica?
- Agregación de datos masivos: Sumatoria de variables de desempeño general entre 10,000 registros.  
- Normalización de resultados: Cálculo de la media aritmética y aplicación de redondeo para facilitar la interpretación.  

## Lógica del programa
El algoritmo se enfoca en la calificación más global del servicio (`nps -> general`).  
Suma todos los puntajes obtenidos y divide el total entre el número de encuestados (`len(datos)`).  
Incluye una validación para evitar errores de división por cero en caso de que la lista de datos esté vacía.  


````python
def promedio_general_satisfaccion(datos):
    # Variable para ir sumando todas las calificaciones generales
    suma_general = 0
    # Calculamos cuántas personas respondieron en total
    total_personas = len(datos)
    
    # Si la lista está vacía, regresamos 0 para no dividir entre cero
    if total_personas == 0:
        return 0
        
    # Iteramos sobre todos los clientes
    for p in datos:
        # Verificamos que la sección "nps" exista en este registro
        if "nps" in p:
            # Sumamos la calificación general al total acumulado
            suma_general += p["nps"]["general"]
            
    # Calculamos el promedio matemático
    promedio = suma_general / total_personas
    
    # Retornamos el promedio redondeado a dos decimales
    return round(promedio, 2)

#Imprimiendo los resultados

resultado = promedio_general_satisfaccion(datos)
print(f"   Calificación promedio: {resultado} / 10 pts")
````

## Conclusión
Esta es la "calificación final" del negocio.  
Es el termómetro general que resume si la operación completa (comida, servicio, tiempo y precio) está cumpliendo con las expectativas del mercado.  


# Reporte 9. Porcentaje de clientes que volverían

## ¿Qué ponemos en práctica?
- Filtrado booleano: Identificación de registros que cumplen con una condición lógica específica (`True` / `False`).  
- Cálculo de ratios: Transformación de conteos absolutos en porcentajes relativos para informes gerenciales.  

## Lógica del programa
El programa inicializa un contador en cero.  
Al recorrer los datos, verifica si la clave `nps -> volveria` es igual a `True`.  
Al finalizar el ciclo, calcula el porcentaje dividiendo a los clientes que dijeron "Sí" entre el total de encuestados, multiplicando por 100.  



````python
def porcentaje_volverian(datos):
    # Contador para las personas que dijeron que SÍ volverían
    si_volverian = 0
    # Obtenemos el total de personas encuestadas
    total_personas = len(datos)
    
    # Protección para evitar división por cero si no hay datos
    if total_personas == 0:
        return "0%"
        
    # Recorremos cada persona en los datos
    for p in datos:
        # Validamos que exista "nps" y que la respuesta de "volveria" sea True (Verdadera)
        if "nps" in p and p["nps"]["volveria"]:
            # Si es True, sumamos 1 a nuestro contador
            si_volverian += 1
            
    # Calculamos el porcentaje: (los que dijeron sí / el total) multiplicado por 100
    porcentaje = (si_volverian / total_personas) * 100
    
    # Retornamos un texto formateado con el símbolo de porcentaje y 2 decimales
    return f"{round(porcentaje, 2)}%"

#Imprimiendo resultados
print("\n--- 9. Porcentaje de clientes que volverían ---")
resultado = porcentaje_volverian(datos)
print(f"   Índice de retorno: {resultado}")
````


## Conclusión
El porcentaje de retorno es el mejor predictor de la lealtad del cliente.  
Un porcentaje bajo de intención de compra futura es una señal de alerta, incluso si las calificaciones de satisfacción actuales son aceptables.  



# Reporte 10. Cálculo del NPS (Net Promoter Score)

## ¿Qué ponemos en práctica?
- Implementación de fórmulas de negocio: Traducción de una métrica estándar de la industria a código Python.  
- Clasificación condicional: Uso de rangos numéricos para segmentar a los usuarios en categorías de lealtad.  

## Lógica del programa
El algoritmo clasifica a cada encuestado basándose en su probabilidad de recomendación (`nps -> recomendacion`):  
- **Promotores:** Calificaciones de 9 o 10.  
- **Detractores:** Calificaciones de 0 a 6.  
- **Pasivos:** Calificaciones de 7 u 8 (no afectan la resta final).  

Aplica la fórmula:
**NPS = %Promotores - %Detractores**


````python
def calculo_nps(datos):
    # Variables para contar cuántos son promotores y cuántos detractores
    promotores = 0
    detractores = 0
    total_personas = len(datos)
    
    # Si no hay registros, el NPS es 0
    if total_personas == 0:
        return 0
        
    # Recorremos la lista de personas
    for p in datos:
        # Validamos que el bloque "nps" exista
        if "nps" in p:
            # Extraemos la calificación de recomendación que dio el cliente
            recomendacion = p["nps"]["recomendacion"]
            
            # Si la calificación es 9 o 10, es un promotor
            if recomendacion >= 9:
                promotores += 1
            # Si la calificación es 6 o menos, es un detractor (los de 7 y 8 son pasivos y se ignoran en esta parte)
            elif recomendacion <= 6:
                detractores += 1
                
    # Calculamos qué porcentaje del total representan los promotores
    porcentaje_promotores = (promotores / total_personas) * 100
    # Calculamos qué porcentaje del total representan los detractores
    porcentaje_detractores = (detractores / total_personas) * 100
    
    # La fórmula oficial del NPS es: % Promotores - % Detractores
    nps = porcentaje_promotores - porcentaje_detractores
    
    # Retornamos el resultado redondeado a dos decimales
    return round(nps, 2)

#imprimiendo resultados
print("\n--- 10. Cálculo del NPS (Net Promoter Score) ---")
resultado = calculo_nps(datos)
print(f"   Puntaje NPS: {resultado}")
print("   (Fórmula: %Promotores - %Detractores)")
````

## Conclusión
El NPS es la métrica de oro para medir la recomendación boca a boca.  
Un NPS positivo indica crecimiento orgánico, mientras que uno negativo sugiere que el restaurante está ganando una mala reputación que debe ser corregida


# Reporte 11. Segmentación: promotores, pasivos, detractores

### ¿Que analiza el reporte?


### ¿Qué ponemos en práctica?
- Uso de diccionarios anidados
- Recorrido de datos con ciclos (for)
- Uso de condicionales (if, elif, else)
- Aplicación del modelo NPS (Net Promoter Score)
- Retorno de datos usando un diccionario

### Logica del programa
El programa recibe una lista de encuestados (datos) y se encarga de clasificarlos según su nivel de recomendación, usando el modelo NPS (Net Promoter Score).
Primero, inicializa tres contadores: promotores, pasivos y detractores, que representan los diferentes tipos de clientes.
Luego recorre cada registro de la lista y obtiene la calificación de recomendación desde la estructura "nps". Dependiendo del valor obtenido, clasifica al cliente:
- Si la calificación es mayor o igual a 9, se considera promotor. 
- Si es 7 u 8, se considera pasivo. 
- Si es menor a 7, se considera detractor. 
A medida que recorre los datos, va sumando en cada categoría correspondiente.
Finalmente, devuelve un diccionario con la cantidad total de promotores, pasivos y detractores, permitiendo así analizar la segmentación de clientes de forma clara.


````python
#Define la función que recibe la lista de encuestados.
def segmentacion_clientes(datos):
    
    #Inicializa contadores para cada tipo de cliente.
    promotores = 0 
    pasivos = 0
    detractores = 0
    
    #Recorre cada encuestado (cada fila es un diccionario).
    for fila in datos:
        
        #Obtiene la calificación de recomendación (pregunta NPS).
        calificacion = fila["nps"]["recomendacion"]
        if calificacion >= 9:
            promotores += 1
        elif calificacion >= 7:
            pasivos += 1
        else:
            detractores += 1
            
    #Retorna los resultados en un diccionario para poder acceder fácilmente a cada valor.
    return {
    "promotores": promotores,
    "pasivos": pasivos,
    "detractores": detractores
    }

#Esta funcion retorna el diccionario y lo almacenamos en una variable
resultado = segmentacion_clientes(datos)

#Imprimimos los resultados por separado
print(f"Promotores: {resultado['promotores']}")
print(f"Pasivos: {resultado['pasivos']}")
print(f"Detractores: {resultado['detractores']}")
````

### Conclusion
Este reporte permite identificar el nivel de satisfacción de los clientes mediante el modelo NPS. Esto es útil para tomar decisiones, ya que muestra cuántos clientes están satisfechos, neutrales o insatisfechos.

# Reporte 14. Relacion de gasto y satisfacción

### ¿Qué ponemos en práctica?
- Uso de diccionarios para organizar la información en grupos (por nivel de gasto).
- Recorrido de listas de encuestados usando ciclos (for).
- Cálculo de promedios de satisfacción usando suma de valores.
- Uso de condicionales (if, elif, ifelse) para clasificar a los clientes.
- Relación entre dos variables importantes: gasto y satisfacción.

### Lógica del programa
El programa toma la información de cada encuestado y obtiene dos datos principales: el gasto del cliente y su nivel de satisfacción (producto y servicio).

Primero calcula un promedio de satisfacción por persona, uniendo ambas calificaciones.

Luego clasifica a los clientes según su gasto en tres grupos:
- Bajo
- Medio
- Alto

Mientras recorre los datos, va acumulando la satisfacción total y el gasto total de cada grupo, además de contar cuántas personas hay en cada uno.

Al final, calcula los promedios de cada grupo dividiendo las sumas entre la cantidad de personas. Esto permite analizar la relación entre gasto y satisfacción.


````python
def relacion_gasto_satisfaccion(datos):
    
    # Diccionario para agrupar por rangos de gasto
    segmentos = {
        "bajo": [0, 0, 0],      # suma_satisfaccion, cantidad, gasto_total
        "medio": [0, 0, 0],
        "alto": [0, 0, 0]
    }

    for p in datos:
        
        # Obtener datos
        gasto = p["consumo"]["gasto"]
        producto = p["experiencia"]["producto"]
        servicio = p["experiencia"]["servicio"]
        
        # Promedio de satisfacción
        satisfaccion = (producto + servicio) / 2
        
        # Clasificación por gasto
        if gasto < 50:
            segmento = "bajo"
        elif gasto <= 100:
            segmento = "medio"
        else:
            segmento = "alto"
        
        # Acumular datos
        segmentos[segmento][0] += satisfaccion
        segmentos[segmento][1] += 1
        segmentos[segmento][2] += gasto

    # Calcular promedios finales
    resultado = {}

    for key, valores in segmentos.items():
        suma_sat = valores[0]
        cantidad = valores[1]
        suma_gasto = valores[2]

        if cantidad > 0:
            resultado[key] = {
                "promedio_satisfaccion": round((suma_sat / cantidad),2),
                "promedio_gasto": round((suma_gasto / cantidad), 2)
            }
        else:
            resultado[key] = {
                "promedio_satisfaccion": 0,
                "promedio_gasto": 0
            }

    return resultado

# Imprimiendo 
print("\n14. Relación entre gasto y satisfacción")

resultado_14 = relacion_gasto_satisfaccion(datos)

for segmento, valores in resultado_14.items():
    print(f"\nSegmento: {segmento}")
    print(f"Promedio de satisfacción: {valores['promedio_satisfaccion']}")
    print(f"Promedio de gasto: {valores['promedio_gasto']}")
````

### Conclusión
Este reporte sirve para entender si el dinero que gasta un cliente influye en qué tan satisfecho está.

Esto es importante porque ayuda a la empresa a saber si los clientes que gastan más están más contentos o no. Con esta información se pueden tomar decisiones para mejorar el servicio, ajustar precios o cambiar estrategias para mejorar la experiencia del cliente.

# Reporte 16. Relación entre precio y recomendación

Se utiliza `.get()` para inicializar valores de forma dinámica y evitar errores si la clave no existe.  
Al final, se calcula un promedio de recomendación por cada categoría de precio.

````python
def reporte16(datos):
    #Calcula la suma de recomendaciones según la percepción del precio.
    analisis = {}
    for fila in datos:
        p = fila["experiencia"]["precio"]
        r = fila["nps"]["recomendacion"]
        # Acumula la recomendación por cada categoría de precio
        analisis[p] = analisis.get(p, 0) + r
    return {"suma": analisis}

#Imprimiendo resultados
print("\n--- 16. Relación entre precio y recomendación (Suma de puntos) ---")
r = reporte16(datos)
for categoria, suma in r['suma'].items():
    print(f"   Percepción: {categoria:15} | Suma Recomendación: {suma}")
````

# Reporte 17. Tiempo de entrega vs Satisfacción


## ¿Qué ponemos en práctica?
- Cálculo de promedios  
- Uso de condicionales  
- Diccionarios  

## Lógica del programa
El programa clasifica los registros según el tiempo de entrega (por ejemplo: `Rápido`, `Normal`, `Lento`).  
Luego, calcula el promedio de satisfacción del cliente para cada categoría utilizando acumuladores y contadores.  


````python
def reporte17(datos):
    #"""Promedio de satisfacción general según el tiempo de entrega."""
    tiempos = {}
    conteos = {}
    for fila in datos:
        t = fila["experiencia"]["tiempo de entrega"]
        s = fila["nps"]["general"]
        tiempos[t] = tiempos.get(t, 0) + s
        conteos[t] = conteos.get(t, 0) + 1
    
    # Genera el promedio por cada tiempo (Rápido, Normal, Lento)
    return {t: round(tiempos[t]/conteos[t], 2) for t in tiempos}

#Imprimiendo resultados
print("\n--- 17. Tiempo de entrega vs Satisfacción (Promedios) ---")
tiempos = reporte17(datos)
for tiempo, promedio in tiempos.items():
    print(f"   Tiempo: {tiempo:15} | Satisfacción Promedio: {promedio}")
````

## Conclusión
Se determina cómo afecta el retraso en el servicio al cliente.

- Retorno de datos ordenados

El algoritmo cuenta la frecuencia de cada tipo de comida utilizando un diccionario.

# Reporte 20. Perfil del cliente promedio

## ¿Qué ponemos en práctica?
- Uso de `len()`  
- Condicionales para evitar errores de división por cero  
- Retorno de resultados  

## Lógica del programa
El algoritmo calcula el promedio general de gasto y el promedio de satisfacción del cliente.  
Se asegura de que existan datos antes de realizar divisiones, evitando errores en la ejecución.  


````python
def reporte20(datos):
    #"""Define el perfil promedio: gasto y satisfacción de todos los encuestados."""
    total_gasto = 0
    total_sat = 0
    cantidad = len(datos)
    
    if cantidad == 0: return "Sin datos"
    
    for fila in datos:
        total_gasto += fila["consumo"]["gasto"]
        total_sat += fila["nps"]["general"]
        
    return {
        "gasto_medio": round(total_gasto / cantidad, 2),
        "satisfaccion_media": round(total_sat / cantidad, 2)
    }

#Imprimiendo resultados
print("\n--- 20. Perfil del cliente promedio ---")
perfil = reporte20(datos)
print(f"   Gasto promedio:        ${perfil['gasto_medio']}")
print(f"   Satisfacción promedio: {perfil['satisfaccion_media']}/10")
````

## Conclusión
Este reporte analiza el comportamiento general del cliente: cuánto gasta y qué tan satisfecho está normalmente, proporcionando una visión integral del desempeño del negocio.
